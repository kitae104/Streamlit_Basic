### 센서 데이터 처리하기

```python
import pandas as pd

# 내용과 형태를 확인
df_normal = pd.read_csv('RawData1.csv', header=0)
print(df_normal.head())
print(df_normal.shape)
```

```python
# 컬럼명 변경하기
df_normal = df_normal.rename(columns={'Time (s)': 'Time', 'Linear Acceleration x (m/s^2)': 'X', 'Linear Acceleration y (m/s^2)': 'Y', 'Linear Acceleration z (m/s^2)': 'Z', 'Absolute acceleration (m/s^2)': 'W'})
print(df_normal.columns)
```

```python
# 데이터 크기 조절(950개)
df_normal = df_normal[:950]
```

```python
# 데이터 확인 
# print(df_normal.describe())
print(df_normal.info())
```

```python
# 데이터 추출 
df_normal = df_normal[['Time','X', 'Y', 'Z']]
print(df_normal.head())
```

```python
# 데이터 시각화
import matplotlib.pyplot as plt
plt.plot(df_normal['Time'], df_normal['X'], label='X', color='red')
plt.plot(df_normal['Time'], df_normal['Y'], label='Y', color='green')
plt.plot(df_normal['Time'], df_normal['Z'], label='Z', color='blue')
plt.legend()
plt.show()
```

```python
# 내용과 형태를 확인
df_abnormal = pd.read_csv('RawData2.csv', header=0)
print(df_abnormal.head())
print(df_abnormal.shape)
```

```python
# 컬럼명 변경하기
df_abnormal = df_abnormal.rename(columns={'Time (s)': 'Time', 'Linear Acceleration x (m/s^2)': 'X', 'Linear Acceleration y (m/s^2)': 'Y', 'Linear Acceleration z (m/s^2)': 'Z', 'Absolute acceleration (m/s^2)': 'W'})
print(df_abnormal.columns)
```

```python
# 데이터 크기 조절(950개)
df_abnormal = df_abnormal[:950]
```

```python
# 데이터 확인 
# print(df_normal.describe())
print(df_abnormal.info())
```

```python
# 데이터 추출 
df_abnormal = df_abnormal[['Time','X', 'Y', 'Z']]
print(df_abnormal.head())
```

```python
# 데이터 시각화
import matplotlib.pyplot as plt
plt.plot(df_abnormal['Time'], df_abnormal['X'], label='X', color='red')
plt.plot(df_abnormal['Time'], df_abnormal['Y'], label='Y', color='green')
plt.plot(df_abnormal['Time'], df_abnormal['Z'], label='Z', color='blue')
plt.legend()
plt.show()
```

```python
# 산점도 그리기
plt.figure(figsize=(8, 6))  # 그래프 크기 조절
plt.scatter(df_normal['X'], df_normal['Y'], label='정상', color='blue', s=10)  # 정상 데이터
plt.scatter(df_abnormal['X'], df_abnormal['Y'], label='비정상', color='red', s=10)  # 비정상 데이터

# 축 레이블 및 제목 설정
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.title('산점도 그래프')

# 그리드 추가 (선택사항)
plt.grid(True)

plt.show()
```

```python
# 산점도 그리기
plt.figure(figsize=(8, 6))  # 그래프 크기 조절
plt.scatter(df_normal['X'], df_normal['Z'], label='정상', color='blue', s=10)  # 정상 데이터
plt.scatter(df_abnormal['X'], df_abnormal['Z'], label='비정상', color='red', s=10)  # 비정상 데이터

# 축 레이블 및 제목 설정
plt.xlabel('x 값')
plt.ylabel('y 값')
plt.title('산점도 그래프')

# 그리드 추가 (선택사항)
plt.grid(True)

plt.show()
```

```python
# 산점도 그리기
plt.figure(figsize=(8, 6))  # 그래프 크기 조절
plt.scatter(df_normal['Y'], df_normal['Z'], label='정상', color='blue', s=10)  # 정상 데이터
plt.scatter(df_abnormal['Y'], df_abnormal['Z'], label='비정상', color='red', s=10)  # 비정상 데이터

# 축 레이블 및 제목 설정
plt.xlabel('Y 값')
plt.ylabel('Z 값')
plt.title('산점도 그래프')

# 그리드 추가 (선택사항)
plt.grid(True)

plt.show()
```

```python
# "Result" 컬럼 추가 및 "YES" / "NO" 값 입력
df_normal['Result'] = 'YES'
print(df_normal.head())
df_abnormal['Result'] = 'NO'
print(df_abnormal.head())
```

```python
# 데이터 프레임 병합하기
df_merged = pd.concat([df_normal, df_abnormal], axis=0, ignore_index=True)
print(df_merged.head())
print(df_merged.shape)
```

```python
# 내용 임의로 섞기 
df_shuffled = df_merged.sample(frac=1)
print(df_shuffled.head())
```
