위의 막대 그래프는 인천터미널 역과 귤현 역의 승차 및 하차 인원을 비교합니다. 인천터미널은 전반적으로 이용 인원이 많고 귤현은 상대적으로 적은 이용 인원을 보여줍니다.

```python
import matplotlib.pyplot as plt

# Convert the '일자' column to datetime format
data['일자'] = pd.to_datetime(data['일자'])

# Group by date to get daily totals
daily_totals = data.groupby('일자').sum()

# Plotting the daily usage trends
plt.figure(figsize=(12, 6))
plt.plot(daily_totals.index, daily_totals['이용인원'], label='Total Usage', marker='o')
plt.plot(daily_totals.index, daily_totals['승차인원'], label='Total Boarding', marker='o')
plt.plot(daily_totals.index, daily_totals['하차인원'], label='Total Alighting', marker='o')

plt.title('Daily Trends of Subway Usage')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Group by date to get daily totals
daily_totals = data.groupby('일자').sum(numeric_only=True)

# Plotting the daily usage trends
plt.figure(figsize=(12, 6))
plt.plot(daily_totals.index, daily_totals['이용인원'], label='Total Usage', marker='o')
plt.plot(daily_totals.index, daily_totals['승차인원'], label='Total Boarding', marker='o')
plt.plot(daily_totals.index, daily_totals['하차인원'], label='Total Alighting', marker='o')

plt.title('Daily Trends of Subway Usage')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
