- csv 모듈의 reader 함수를 이용하여 행별로 데이터를 가져와 루프를 돌면 csv 파일을 읽는다.
- next 함수는 다음 행으로 이동하고 싶거나 행을 넘겨버리고 싶을 때 사용한다.
- 처음 open 함수를 이용하여 파일을 열었다면 마지막에는 close 함수를 사용하여 파일을 닫아줘야 한다.
- matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음이다.
- matplotlib.pyplot 모듈은 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있다.

```python
import csv

a = open("./datas/인천교통공사_1호선 일별 승하차현황_20240630.csv") # cp949
data = csv.reader(a)    # csv.reader() 함수를 이용하여 파일을 읽어옴

for row in data:
    print(row)      # 리스트 형태로 출력
a.close()
```
