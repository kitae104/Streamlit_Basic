```python
import pandas as pd
import seaborn as sns 

df = sns.load_dataset('penguins')
df.head()
```

```python
# query 메서드를 이용한 부등호 필터링
df.query('bill_length_mm > 55')

# 위 코드와 동일한 결과를 얻음
# df[df['bill_length_mm'] > 55]

# 위 코드와 동일한 결과를 얻음
# df.loc[df['bill_length_mm'] > 55]
```

```python
# query 메서드를 이용한 and 조건문 연결
df.query('bill_length_mm > 55 and species == "Gentoo"')
```

```python
# query 메서드 조건문의 외부 변수 참조
length = 55
species = 'Chinstrap'

df.query('bill_length_mm >= @length and species == @species')
```

```python
# query 메서드 조건문의 문자열 메서드 사용
df.query('island.str.contains("oe")', engine='python')  # engine='python' 옵션을 사용해 문자열 메서드를 사용할 수 있음

df.query('species.str.endswith("e")', engine='python')  # species 열의 값이 'e'로 끝나는 행을 필터링
```

```python
# isin을 이용한 리스트 내 항목 참조
filtering = ["Adelie", "Chinstrap"]
df.query('species.isin(@filtering)', engine='python')  # species 열의 값이 'Adelie' 또는 'Chinstrap'인 행을 필터링
```
